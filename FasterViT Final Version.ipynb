```python
import torch
from torchvision import transforms
from PIL import Image
from fastervit import create_model

# Define fastervit-0 model with 224 x 224 resolution

# Step 1: Load the pre-trained FasterViT model
model = create_model('faster_vit_0_224', 
                          pretrained=True,
                          model_path="/tmp/faster_vit_0.pth.tar")
model.eval()

num_ftrs = model.head.in_features
model.head = torch.nn.Linear(num_ftrs, 8)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
model
```

```python
import torch.optim as optim
import torch.nn as nn

criterion = nn.CrossEntropyLoss()  # or any other suitable loss function
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

# Started from 6/25/2024 5:48 PM

```python
num_epochs = 12  # Set the number of epochs

for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0
    print(epoch)

    for images, labels in train_loader:
        optimizer.zero_grad()  # Zero the parameter gradients

        outputs = model(images)  # Forward pass
        loss = criterion(outputs, labels)  # Compute the loss
        loss.backward()  # Backward pass
        optimizer.step()  # Optimize the parameters

        running_loss += loss.item()

    # Print training statistics
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader)}")

    # Validation phase
    model.eval()  # Set the model to evaluation mode
    val_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    # Print validation statistics
    print(f"Validation Loss: {val_loss/len(val_loader)}, Accuracy: {100 * correct / total}%")
```

```python
# Create datasets
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Create dataloaders
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
model.eval()
```

```python
# Define a loss function and other necessary components
criterion = torch.nn.CrossEntropyLoss()

# Testing loop
test_loss = 0.0
correct = 0
total = 0

with torch.no_grad():  # No need to calculate gradients during inference
    for data, targets in test_loader:
        outputs = model(data)
        loss = criterion(outputs, targets)
        test_loss += loss.item() * data.size(0)
        _, predicted = torch.max(outputs, 1)
        total += targets.size(0)
        correct += (predicted == targets).sum().item()

# Calculate average loss and accuracy
test_loss = test_loss / len(test_loader.dataset)
accuracy = 100 * correct / total

print(f'Test Loss: {test_loss:.4f}, Test Accuracy: {accuracy:.2f}%')
```

```python
# Collect true and predicted labels
all_labels = []
all_preds = []

with torch.no_grad():  # No need to calculate gradients during inference
    for data, targets in test_loader:
        outputs = model(data)
        _, predicted = torch.max(outputs, 1)
        all_labels.extend(targets.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())
```

```python
# Compute MCC
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(all_labels, all_preds)
print(f'Matthews Correlation Coefficient: {mcc:.4f}')
```
