```python
y_pred_vit = loaded_model.predict(X_test_float32)
```

```python
y_pred_vit_without_one_hot = np.argmax(y_pred_vit, axis=1)
```

```python
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import confusion_matrix
from sklearn.metrics import *
```

```python
print("Matthews Correlation : ",matthews_corrcoef(y_val, y_pred_vit_without_one_hot))
print("Confusion Matrix : \n",confusion_matrix(y_val, y_pred_vit_without_one_hot))
print("Accuracy on test set:   ",accuracy_score(y_val,y_pred_vit_without_one_hot))
```
